```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# RTT = 4 #
p = [0.1, 0.15, 0.2, 0.25, 0.3]
exp = [0.87, 0.76, 0.67, 0.58, 0.51]
theor = [0.96, 0.79, 0.68, 0.61, 0.55]

#sns.set_style("whitegrid")
plt.xlabel('p ошибки')
plt.ylabel('ƞ')
plt.title('Зависимость коэфф. использования канала от вероятности ошибки при RTT = 4')
plt.grid(False)
plt.plot(p, exp, 'purple')
plt.plot(p, theor, 'pink')

plt.legend(['Практическое', 'Теоретическое'])
plt.show()
```
